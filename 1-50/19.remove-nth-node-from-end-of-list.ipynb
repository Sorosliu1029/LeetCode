inspired by hint, two pointers, delay one, use dummy to ease special condition

```python
from typing import Optional

# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
class Solution:
    def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
        dummy = ListNode(0, next=head)
        lead, delay = head, dummy
        cnt = 0
        while lead:
            cnt += 1
            if cnt > n:
                delay = delay.next
            lead = lead.next

        delay.next = delay.next.next
        return dummy.next
```
