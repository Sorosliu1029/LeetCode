### 621. Task Scheduler

<p>You are given an array of CPU <code>tasks</code>, each labeled with a letter from A to Z, and a number <code>n</code>. Each CPU interval can be idle or allow the completion of one task. Tasks can be completed in any order, but there&#39;s a constraint: there has to be a gap of <strong>at least</strong> <code>n</code> intervals between two tasks with the same label.</p>

<p>Return the <strong>minimum</strong> number of CPU intervals required to complete all tasks.</p>

---
What's your idea?

greedy, maximize gap between same task, so choose task by highest remaining count, check if meet interval requirement.
If so, schedule it, if not, check next highest

---

```python
from typing import List
```

```python
import heapq

class Solution:
    def leastInterval(self, tasks: List[str], n: int) -> int:
        cnts = [0]*26
        for t in tasks:
            cnts[ord(t)-ord('A')] += 1
        pq = [(-c, chr(i+ord('A'))) for i, c in enumerate(cnts) if c > 0]
        heapq.heapify(pq)

        delays = {}
        total = 0
        while pq:
            re_heap = []
            scheduled = None
            while pq:
                remain, task = heapq.heappop(pq)
                if remain == 0:
                    continue
                if delays.get(task, 0) != 0:
                    re_heap.append((remain, task))
                else:
                    remain += 1
                    re_heap.append((remain, task))
                    delays[task] = n
                    scheduled = task
                    break
                
            pq.extend(re_heap)
            heapq.heapify(pq)
            
            total += 1
            for k, v in delays.items():
                if v != 0 and k != scheduled:
                    delays[k] -= 1

        return total-1
```
