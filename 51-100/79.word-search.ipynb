### 79. Word Search

<p>Given an <code>m x n</code> grid of characters <code>board</code> and a string <code>word</code>, return <code>true</code> <em>if</em> <code>word</code> <em>exists in the grid</em>.</p>

<p>The word can be constructed from letters of sequentially adjacent cells, where adjacent cells are horizontally or vertically neighboring. The same letter cell may not be used more than once.</p>

What's your idea?

find all the start points, dfs from each start point, use stack to record ongoing match and visited map

---

```python
from typing import List
```

```python
class Solution:
    def exist(self, board: List[List[str]], word: str) -> bool:
        m = len(board)
        n = len(board[0])
        target_p = len(word)-1
        neighbours = [(0,-1), (0,1), (-1,0), (1,0)] # left, right, up, down
        starts = []
        
        for i in range(m):
            for j in range(n):
                if board[i][j] == word[0]:
                    # i, j, and map of visited with (i,j):0 to record index of word
                    starts.append((i,j, {(i,j):0}))
                    
        for i, j, visited in starts:
            stack = [(i,j, visited)]
            while stack:
                curri, currj, visited = stack.pop()
                visited_p = visited.get((curri, currj))
                if visited_p == target_p:
                    return True
                for di, dj in neighbours:
                    ni, nj = curri+di, currj+dj
                    if 0 <= ni < m and 0 <= nj < n and visited.get((ni,nj)) is None and board[ni][nj] == word[visited_p+1]:
                        copy_visited = dict(visited)
                        copy_visited[(ni,nj)] = visited_p+1
                        stack.append((ni, nj, copy_visited))

        return False
```
