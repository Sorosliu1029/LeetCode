```python
class Solution:
    def reverseBits(self, n: int) -> int:
        m = 0
        for i in range(31):
            m |= (1 if n % 2 else 0)
            n >>= 1
            m <<= 1
        return m
```

```python
s = Solution()
s.reverseBits(43261596)
```

```python
s.reverseBits(2147483644)
```
