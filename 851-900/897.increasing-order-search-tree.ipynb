### 897. Increasing Order Search Tree

#### Content
<p>Given the <code>root</code> of a binary search tree, rearrange the tree in <strong>in-order</strong> so that the leftmost node in the tree is now the root of the tree, and every node has no left child and only one right child.</p>

```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
class Solution:
    def increasingBST(self, root: TreeNode) -> TreeNode:
        root, _ = self.recur(root)
        return root

    def recur(self, node):
        if node is None:
            return None, None
        left_list_head, left_list_tail = self.recur(node.left)
        right_list_head, right_list_tail = self.recur(node.right)
        if not left_list_tail:
            left_list_head = node
        else:
            left_list_tail.right = node
        node.right = right_list_head
        node.left = None
        if not right_list_tail:
            right_list_tail = node
        return left_list_head, right_list_tail
```

```python
from helper import array_to_tree
from visualizer import visualize_binary_tree
s = Solution()

t = array_to_tree([1])
visualize_binary_tree(s.increasingBST(t))

t = array_to_tree([5,1])
visualize_binary_tree(s.increasingBST(t))

t = array_to_tree([5,1,7])
visualize_binary_tree(s.increasingBST(t))

t = TreeNode(5,
  TreeNode(3,
    TreeNode(2,
      TreeNode(1)
    ),
    TreeNode(4),
  ),
  TreeNode(6,
    None,
    TreeNode(8,
      TreeNode(7),
      TreeNode(9)
    ),
  ),
)
visualize_binary_tree(s.increasingBST(t))
```
